```python
class Student:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.finished_courses = []
        self.courses_in_progress = []
        self.grades = {}
        self.ever_grades = {}
        self.average = []
        
    def rate_lec(self, lecturer, course, grade):
        if (lecturer, Lecturer) and course in self.courses_in_progress and course in lecturer.courses_attached:
            if course in lecturer.grades:
                lecturer.grades[course] += [grade]
            else:
                lecturer.grades[course] = [grade]
        else:
            return 'Ошибка'
            
    def grades_lec(self, lecturer, course):
        print('Enter rating')
        a = True
        grade = 0
        while a:
            grade = int(input())
            if (lecturer, Lecturer) and course in self.courses_in_progress and course in lecturer.courses_attached:
                if course in lecturer.grades:
                    lecturer.grades[course] += [grade]
                else:
                    lecturer.grades[course] = [grade]
            else:
                return 'Ошибка'
            a = input('Do you want to continue, yes of no?: ')
            if a == 'yes':
                a = True
            else:
                break
    def ever_grade_course(self, course):
        average = 0
        count = 0
        for i in self.grades[course]:
            average += i
            count += 1
        self.ever_grades[course] = round(average / count, 1)
          
    def ever_grade(self):
        average = 0
        for i in self.ever_grades:
            average += self.ever_grades[i]
        self.average = round(average / len(self.ever_grades), 1)

    def __lt__(self, other):
        if not isinstance(other, Student):
            print('Not a Student!')
            return
        return self.average < other.average
    
    def __str__ (self):
        return str(f'Имя: {self.name}\n'
        f'Фамилия: {self.surname}\n'
        f'Средняя оценка за домашние задания: {self.average}\n'
        f'Курсы в процессе изучения: {", ".join(self.courses_in_progress)}\n'
        f'Завершенные курсы: {", ".join(self.finished_courses)}\n')

        
class Mentor:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.courses_attached = []
        self.grades = {}
        self.ever_grades = {}
        self.average = []
        
class Lecturer(Mentor):
    def __str__(self):
        return str(f'Имя: {self.name}\n'
        f'Фамилия: {self.surname}\n'
        f'Средняя оценка за лекции: {self.average}\n')

    def __lt__(self, other):
        if not isinstance(other, Lecturer):
            print('Not a Student!')
            return
        return self.average < other.average
    
    def ever_grade_course(self, course):
        average = 0
        count = 0
        for i in self.grades[course]:
            average += i
            count += 1
        self.ever_grades[course] = round(average / count, 1)

    def ever_grade(self):
        average = 0
        for i in self.ever_grades:
            average += self.ever_grades[i]
        self.average = round(average / len(self.ever_grades), 1)

class Reviewer(Mentor):
    def __str__(self):
        return str(f'Имя: {self.name}\n'
        f'Фамилия: {self.surname}\n')
        
    def rate_hw(self, student, course, grade):
        if (student, Student) and course in self.courses_attached and course in student.courses_in_progress:
            if course in student.grades:
                student.grades[course] += [grade]
            else:
                student.grades[course] = [grade]
        else:
            return 'Ошибка'
    
    def grades_hw(self, student, course):
        a = True
        grade = 0
        while a:
            grade = int(input('Enter rating: '))
            if (student, Student) and course in self.courses_attached and course in student.courses_in_progress:
                if course in student.grades:
                    student.grades[course] += [grade]
                else:
                    student.grades[course] = [grade]
            else:
                return 'Ошибка'
            a = input('Do you want to continue, yes of no?: ')
            if a == 'yes':
                a = True
            else:
                break
                
def average_students(list, course):
    count = 0
    average = 0
    for i in list:
        average += i.ever_grades[course]
        count += 1
    return round(average / count, 1)
            
def average_lecturers(list, course):
    count = 0
    average = 0
    for i in list:
        average += i.ever_grades[course]
        count += 1
    return round(average / count, 1)
    
students = []
lecturers = []

student1 = Student('Bobby', 'Dylan', 'your_gender')
student1.courses_in_progress += ['Python']
student1.courses_in_progress += ['Git']
student1.finished_courses += ['Введение в программирование']
students += [student1]

student2 = Student('Adam', 'Levine', 'your_gender')
student2.courses_in_progress += ['Python']
student2.courses_in_progress += ['Git']
student2.finished_courses += ['Введение в программирование']
students += [student2]

rew1 = Reviewer('Some', 'Buddy', 'your_gender')
rew1.courses_attached += ['Python']
# rew1.courses_attached += ['Git']

rew2 = Reviewer('Ben', 'Mee', 'your_gender')
# rew2.courses_attached += ['Python']
rew2.courses_attached += ['Git']

lec1 = Lecturer('Billie', 'Jean', 'your_gender')
lec1.courses_attached += ['Python']
lec1.courses_attached += ['Git']
lecturers += [lec1]

lec2 = Lecturer('Marco', 'Polo', 'your_gender')
lec2.courses_attached += ['Python']
lec2.courses_attached += ['Git']
lecturers += [lec2]

rew1.rate_hw(student1, 'Python', 5)
rew1.rate_hw(student1, 'Python', 9)
rew1.rate_hw(student1, 'Python', 7)
rew2.rate_hw(student1, 'Git', 9)
rew2.rate_hw(student1, 'Git', 7)
rew2.rate_hw(student1, 'Git', 8)

rew1.rate_hw(student2, 'Python', 8)
rew1.rate_hw(student2, 'Python', 5)
rew1.rate_hw(student2, 'Python', 5)
rew2.rate_hw(student2, 'Git', 6)
rew2.rate_hw(student2, 'Git', 8)
rew2.rate_hw(student2, 'Git', 5)

student1.ever_grade_course('Python')
student1.ever_grade_course('Git')
student1.ever_grade()

student2.ever_grade_course('Python')
student2.ever_grade_course('Git')
student2.ever_grade()

student1.rate_lec(lec1, 'Git', 9)
student1.rate_lec(lec1, 'Git', 8)
student1.rate_lec(lec1, 'Git', 10)
student1.rate_lec(lec2, 'Python', 9)
student1.rate_lec(lec2, 'Python', 9)
student1.rate_lec(lec2, 'Python', 8)

student2.rate_lec(lec1, 'Python', 6)
student2.rate_lec(lec1, 'Python', 6)
student2.rate_lec(lec1, 'Python', 5)
student2.rate_lec(lec2, 'Git', 5)
student2.rate_lec(lec2, 'Git', 10)
student2.rate_lec(lec2, 'Git', 5)

lec1.ever_grade_course('Python')
lec1.ever_grade_course('Git')
lec1.ever_grade()

lec2.ever_grade_course('Python')
lec2.ever_grade_course('Git')
lec2.ever_grade()

print(student1.ever_grades)
print(student2.ever_grades)
# print(lec1.ever_grades)
# print(lec2.ever_grades)
# print(rew1)
# print(rew2)
print(average_students(students, 'Python'))
print(average_students(students, 'Git'))
# print(average_lecturers(lecturers, 'Python'))
# print(average_lecturers(lecturers, 'Git'))
```
